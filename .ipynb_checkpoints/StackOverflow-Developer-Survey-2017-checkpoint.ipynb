```python
#importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
%matplotlib inline
```

## 2-Data Gathering and Understanding:

```python
#loading dataset
df=pd.read_csv('./survey_results_public.csv')
df.head(5)
```

```python
#finding number of rows and columns
df.shape
```

```python
#finding name and number of columns with zero missing value
zero_miss_col=df.columns[df.isnull().sum()==0]
print(zero_miss_col)
print(zero_miss_col.shape[0])
```

```python
#finding name and number of columns with greater than 50% missing value
large_miss_col=df.columns[(df.isnull().sum()/df.shape[0])*100>50]
print(large_miss_col)
print(large_miss_col.shape[0])
```

```python
#finding name and number of columns with all missing value
all_miss_col=df.columns[df.isnull().sum()==df.shape[0]]
print(all_miss_col)
print(all_miss_col.shape[0])
```

```python
#finding name and number of quantitative columns
num_df=df.select_dtypes(include=['int','float'])
num_col=num_df.columns
print(num_col)
print(num_col.shape[0])
```

```python
#finding name and number of categorical columns
cat_df=df.select_dtypes(include=['object']).copy()
cat_col=cat_df.columns
print(cat_col)
print(cat_col.shape[0])
```

## 3-Data Preparation:

### 3.1-Dropping all the rows from the dataframe containing 'Not Employed' in their 'Employmnet Status column                 as all the above business question are interested in the respondents who are employed.

```python
#dropping all rows from the dataframe containing 'Not Employed' in their 'Employmnet Status column
emp_df=df[(df['EmploymentStatus']=='Employed full-time') | (df['EmploymentStatus']=='Employed part-time')]
emp_df
```

### 3.2-Imputing NaN value in the quantitative columns with their mean value

```python
#imputing NaN value in the quantitative columns with their mean value
for col in num_col:
    emp_df[col].fillna(emp_df[col].mean(),inplace=True)
    
emp_df
```

### 3.3- Imputing NaN value in Categorical column with their mode value.

```python
#Imputing NaN value in Categorical column with their mode value.
for col in cat_col:
    emp_df[col].fillna(emp_df[col].mode()[0],inplace=True)
    
emp_df
```

### 4.3:Analysis of Question 3:-Which is the best website or platform to upload your profile or resume?

```python
#Analysis of question 3
platform_count=emp_df.JobProfile.value_counts()
platform_count
```

```python
def total_count(df, col1, col2, look_for):
    '''
    INPUT:
    df - the pandas dataframe you want to search
    col1 - the column name you want to look through
    col2 - the column you want to count values from
    look_for - a list of strings you want to search for in each row of df[col]

    OUTPUT:
    new_df - a dataframe of each look_for with the count of how often it shows up
    '''
    new_df = defaultdict(int)
    #loop through list of ed types
    for val in look_for:
        #loop through rows
        for idx in range(df.shape[0]):
            #if the ed type is in the row add 1
            if val in df[col1][idx]:
                new_df[val] += int(df[col2][idx])
    new_df = pd.DataFrame(pd.Series(new_df)).reset_index()
    new_df.columns = [col1, col2]
    new_df.sort_values('count', ascending=False, inplace=True)
    return new_df
```

```python
possible_vals = ["LinkedIn","Xing","Indeed","Glassdoor","Hired.com","TripleByte","Monster","Dice","JobSite.co.uk","Talent.io","Reed.co.uk","Total_Jobs","Remix_jobs",
                 "StepStone","Naukri","Pracuj","Other"]

def clean_and_plot(df, title='Platform For Maintaining Profile,Resume', plot=True):
    '''
    INPUT 
        df - a dataframe holding the CousinEducation column
        title - string the title of your plot
        axis - axis object
        plot - bool providing whether or not you want a plot back
        
    OUTPUT
        study_df - a dataframe with the count of how many individuals
        Displays a plot of pretty things related to the CousinEducation column.
    '''
    platform = df['JobProfile'].value_counts().reset_index()
    platform.rename(columns={'index': 'platform', 'JobProfile': 'count'}, inplace=True)
    platform_df = total_count(platform, 'platform', 'count', possible_vals)

    platform_df.set_index('platform', inplace=True)
    if plot:
        (platform_df/platform_df.sum()).plot(kind='bar', legend=None);
        plt.title(title);
        plt.show()
    props_platform_df = platform_df/platform_df.sum()
    return props_platform_df
    
props_df = clean_and_plot(emp_df)
```

### Result:
From the above plot, we can see that more than 70% of the employed software developer maintains their profile,cv or resume on Indeed.So,Indeed is the best platform to maintains profile or resume for job searching.
